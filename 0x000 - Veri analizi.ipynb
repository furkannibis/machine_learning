```python
# Eksik verileri tamamlama
df = pd.read_csv("http://bilkav.com/eksikveriler.csv")

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="mean")

Yas = df.iloc[:,3:4].values
imputer = imputer.fit_transform(Yas)
df.iloc[:,3:4] = imputer
df
```

```python

# Sözel verileri sayısal veriye çevirme

# Label Encoding
cinsiyet = df.iloc[:,4:5].values

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
cinsiyet = le.fit_transform(cinsiyet)
df.iloc[:,4:5] = cinsiyet
df
```

```python
# Metinsel veriyi sayısal veriye çevirme

# One Hot Encoder
ulke = df.iloc[:,0:1].values

from sklearn import preprocessing
ohe = preprocessing.OneHotEncoder()

ulke = ohe.fit_transform(ulke).toarray()
ulke
```

```python
sonuc = pd.DataFrame(data=ulke, index = range(len(ulke)), columns=["fr", "tr", "usa"])
sonuc
```

```python
sonuc_1 = df.iloc[:,1:]
sonuc_1
```

```python
result_1 = pd.concat([sonuc,sonuc_1], axis=1)
result_1
```

```python
data = result_1.to_csv('veriler.csv', index=False)
```
